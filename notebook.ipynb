# 2. Data Aggregation & Merging

```python
support_agg = customer_support_df.groupby('user_id').agg(
    support_tickets_count=('ticket_time', 'count'),
    avg_resolution_hours=('resolution_time_hours', 'mean')
).reset_index()
```

```python
activity_agg = user_activity_df.groupby('user_id').agg(
    total_activities=('event_time', 'count')
).reset_index()
```

```python
support_agg = support_agg.rename(columns={'user_id': 'customer_id'})
activity_agg = activity_agg.rename(columns={'user_id': 'customer_id'})
```

```python
df_final = account_info_df.merge(support_agg, on='customer_id', how='left')
df_final = df_final.merge(activity_agg, on='customer_id', how='left')
```

```python
df_final['support_tickets_count'] = df_final['support_tickets_count'].fillna(0)
df_final['avg_resolution_hours'] = df_final['avg_resolution_hours'].fillna(0)
df_final['total_activities'] = df_final['total_activities'].fillna(0)
```

## Conclusion

To identify churn drivers, I aggregated the event-level data to the customer level (one row per customer):
*   **Support:** Grouped by user ID to calculate `support_tickets_count` and `avg_resolution_hours`.
*   **Engagement:** Grouped by user ID to calculate `total_activities`.
*   **Merging:** Used a `LEFT JOIN` to combine these metrics with the main `account_info` table. Customers with no activity or support tickets were logically filled with 0s.

# 3. Exploratory Data Analysis (EDA)

```python
sns.set_theme(style="whitegrid")
df_final['churn_label'] = df_final['churn_status'].map({1: 'Churned', 0: 'Active'})
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

churn_counts = df_final['churn_label'].value_counts()
axes[0, 0].pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=90, colors=['#2ca02c', '#d62728'])
axes[0, 0].set_title('Overall Churn Distribution', fontsize=14, fontweight='bold')

sns.countplot(data=df_final, x='plan', order=df_final['plan'].value_counts().index, palette='Blues_r', ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Users by Plan Type', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Number of Users')


sns.barplot(data=df_final, x='plan', y='churn_status', errorbar=None, palette='Reds_r', ax=axes[1, 0])
axes[1, 0].set_title('Churn Rate by Plan Type', fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel('Churn Rate (%)')
axes[1, 0].set_yticklabels(['{:,.0%}'.format(x) for x in axes[1, 0].get_yticks()])


sns.boxplot(data=df_final, x='churn_label', y='total_activities', palette=['#2ca02c', '#d62728'], ax=axes[1, 1])
axes[1, 1].set_title('User Engagement (Total Activities) vs Churn', fontsize=14, fontweight='bold')
axes[1, 1].set_ylabel('Total Activities')
axes[1, 1].set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
print(df_final.groupby('churn_label')[['support_tickets_count', 'avg_resolution_hours']].mean())
```

## Conclusion

**Key EDA Findings:**
1.  **Overall Churn:** The pie chart illustrates that **28.5%** of the user base has already churned, confirming the leadership's concern.
2.  **Plan Distribution:** The user base is evenly distributed across plans, but the **'Free' plan experiences the highest churn rate** (>40%), followed closely by the 'Basic' plan.
3.  **Engagement (Total Activities):** The boxplot demonstrates that **churned users have notably lower total activities** compared to active users. Lack of engagement is a strong predictor of leaving.
4.  **Support Bottlenecks:** Data aggregation shows a massive disparity in support: retained users get their tickets resolved in ~6 hours, while **churned users waited an average of ~17.8 hours**.

# 4. Definition of a Business Metric

**How should the business monitor what they want to achieve?**
The primary business goal is to retain customers. Therefore, the core metric leadership must monitor is the **Customer Churn Rate** (the percentage of subscribers who cancel their subscription over a given period). 

Additionally, based on the exploratory analysis, support performance is a massive driver of churn. The business should implement a secondary operational metric: **Average Support Resolution Time**. Monitoring this will serve as an early warning system; keeping resolution times low will directly prevent future churn.

**Estimate of Initial Values based on current data:**
1. **Initial Churn Rate:** Based on the overall dataset, the current baseline churn rate is **28.5%** (114 churned users out of 400 total users).
2. **Initial Avg. Resolution Time:** Currently, active users have their tickets resolved in **~6.0 hours**, whereas users who ended up churning had to wait an average of **~17.8 hours**. The business should set a KPI to keep all ticket resolution times closer to the 6-hour baseline.

# 5. Final Summary & Recommendations

**Summary of Findings:**
Through our analysis, we identified a high overall churn rate of 28.5%. The data reveals three main drivers of this churn:
1. **Support Bottlenecks:** Users who churn experience ticket resolution times nearly 3 times longer (17.8 hours) than retained users (6.0 hours). 
2. **Low Engagement:** Churned users have significantly fewer total activities in the app compared to active users. If they don't use the app, they don't see the value.
3. **Plan Vulnerability:** The 'Free' plan has the highest churn rate (over 40%), and the 'Basic' plan is also underperforming compared to premium tiers. 

**Recommendations for the Business:**
Based on these findings, I recommend the leadership team undertake the following actions next quarter:

1. **Overhaul Support SLAs (Service Level Agreements):** This is the most urgent fix. Implement strict internal SLAs requiring customer support to resolve technical and account issues in under 8 hours. The data clearly shows that making users wait 17+ hours heavily drives them to cancel.
2. **Launch a Re-engagement Campaign:** Product and Marketing teams should target users who have 0 or 1 total activities. Send automated push notifications, workout reminders, or personalized emails highlighting app features to drive them back into the app before they churn.
3. **Investigate the 'Free' and 'Basic' Onboarding:** Since these plans experience the highest drop-offs, Product Management needs to review the onboarding process for these specific tiers. Consider offering a time-limited "Pro" trial to Free/Basic users so they can experience the full value of the app, encouraging upgrades instead of cancellations.
